```python
import pandas as pd
```

```python
movies = pd.read_csv('u.item.csv', encoding = 'latin1', sep='|', names = ['movie id', 'movie title', 'release date', 'video release date', ' IMDb URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', ' Thriller', 'War', 'Western'])
```

```python
movies.head()
```

```python
movies = movies[['movie id', 'movie title', 'release date', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', ' Thriller', 'War', 'Western']]

movies.head()
```

```python
ratings = pd.read_csv('u.data.csv', sep='\t', names = ['user id', 'item id', 'rating', 'timestamp'])
```

```python
ratings.head()
```

```python
# подсчет количества оценок для каждого пользователя
# сортировка в порядке убывания количества оценок
# получение одной строки с пользователем поставившим наибольшее количество оценок

ratings_count = ratings[['user id', 'rating']].groupby(['user id']).count().sort_values(by = 'rating', ascending=False).head(1)
ratings_count
```

```python
# фильмы по которым пользователь 405 ставил оценки
# и оценки этого пользователя каждому из этих фильмов

ratings_405 = ratings[['item id', 'rating']].loc[ratings['user id'] == 405]
ratings_405
```

```python
# объединение с датафреймом содержащим информацию по фильмам

df = ratings_405.merge(movies, left_on = 'item id', right_on = 'movie id', how = 'inner')
df.head()
```

```python
# добавление колонки с годом

df['year'] = df['release date'].str.split('-').str.get(2)
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[['movie id', 'year', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-Fi', ' Thriller', 'War', 'Western', 'rating']]

df.head()
```

```python
# подсчет общего количества оценок по каждому фильму

ratings_all_count = ratings[['item id', 'rating']].groupby(['item id']).count()
ratings_all_count.head()
```

```python
ratings_all_count.columns = ['rating_count']
ratings_all_count.head()
```

```python
# подсчет общей суммы оценок по каждому фильму

ratings_all_sum = ratings[['item id', 'rating']].groupby(['item id']).sum()

ratings_all_sum.head()
```

```python
ratings_all_sum.columns = ['rating_sum']
ratings_all_sum.head()
```

```python
# получение итогового датафрейма с данными по каждому фильму: 
# год выпуска, жанры, суммарное количество оценок, суммарная оценка от всех пользователей
# а также оценка пользователя 405 по каждому из фильмов

df_result = df.merge(ratings_all_count, left_on = 'movie id', right_on = 'item id', how = 'inner')
df_result = df_result.merge(ratings_all_sum, left_on = 'movie id', right_on = 'item id', how = 'inner')
df_result.head()
```

```python
df_result.describe()
```

```python
df_result.info()
```

```python
df_result['year'] = pd.to_numeric(df_result['year'])
```

```python
df_result.info()
```

```python
df_result.columns
```

```python
X = df_result[['year', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
       'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
       ' Thriller', 'War', 'Western', 'rating_count', 'rating_sum']]

y = df_result['rating']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Разделение на обучающую и тестовую выборки

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
pd.DataFrame([y_test.values, model.predict(X_test)]).T
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Оценка качества модели

mean_squared_error(y_test, model.predict(X_test))
```
